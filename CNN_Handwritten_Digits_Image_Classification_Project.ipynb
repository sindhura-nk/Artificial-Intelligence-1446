```python
from keras.utils import to_categorical
ytrain_scl = to_categorical(ytrain)
ytest_scl = to_categorical(ytest)
```

```python
import numpy as np
xtrain_scl = np.expand_dims(xtrain_scl,axis=-1)
xtest_scl = np.expand_dims(xtest_scl,axis=-1)
```

```python
xtrain_scl.shape
```

## CNN Architechture
Model Building

image.png

```python
ytrain_scl.shape[1]
```

```python
from keras.models import Sequential
from keras.layers import Input,Dense,Conv2D,MaxPool2D,BatchNormalization,Dropout,Flatten

# Initialize the model
model = Sequential()

# Add the INput layer
model.add(Input(shape=(28,28,1),))

# Add the Convolutional Layer
model.add(Conv2D(filters=32,kernel_size=(2,2),activation='relu'))
# Add the Pooling layer
model.add(MaxPool2D(pool_size=(2,2)))

# Add the Convolutional Layer
model.add(Conv2D(filters=64,kernel_size=(2,2),activation='relu'))
# Add the Pooling layer
model.add(MaxPool2D(pool_size=(2,2)))

# Add the Flatten layer
model.add(Flatten()) # convert all the finalised pooled data into 1D data-features

# Add the fully connected layer
model.add(Dense(units=32,activation='relu')) # One hidden layer
model.add(BatchNormalization())
model.add(Dropout(rate=0.2))
model.add(Dense(units=10,activation='softmax')) # Output layer

```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
from keras.callbacks import EarlyStopping
es = EarlyStopping(monitor='val_loss',patience=10)
```

```python
nn = model.fit(xtrain_scl,ytrain_scl,validation_split=0.2,epochs=200,callbacks=[es])
```

## Learning Curve

```python
plt.plot(nn.history['loss'],label='Training Loss')
plt.plot(nn.history['val_loss'],label='Validation Loss')
plt.legend()
plt.xlabel('Number of Epochs')
plt.ylabel('Categorical CrossEntropy')
plt.title('Learning Curve')
plt.show()
```

```python
def preprocess_images(model,img_path):
  # Read the image
  img = cv2.imread(img_path,cv2.IMREAD_GRAYSCALE)
  # Convert the background of image from white to black. Model was trained on black background images
  img = 255-img
  # Scaling the images
  img_scl = img.astype('float32')/255
  # Reshaping images
  img_scl = img_scl.reshape(1,28,28,1)
  # model predictions
  preds = model.predict(img_scl)
  # convert the probabilities into proper predictions
  final_preds = np.argmax(preds,axis=1)
  return final_preds
```
